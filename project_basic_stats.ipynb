```python
import pandas as pd
```

```python
#Numerical data statistics
numerical_stats = df.describe()
# View the numerical statistics
print(numerical_stats)
```

```python
#categorical data statistics
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    unique_count = df[col].nunique()
    print(f"Column '{col}': {unique_count} unique values")
```

```python
#Mode for numerical data
for column in df.select_dtypes(include=['number']):
    print(f"Mode of {column}: {df[column].mode()[0]}")
```

```python
#Findiong missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#Skew and kurtosis


variable1 = 'score_home'
variable2 = 'score_away'

skew1 = df[variable1].skew()
kurtosis1 = df[variable1].kurtosis()

skew2 = df[variable2].skew()
kurtosis2 = df[variable2].kurtosis()

print(f"Skewness of {variable1}: {skew1}")
print(f"Kurtosis of {variable1}: {kurtosis1}")
print(f"Skewness of {variable2}: {skew2}")
print(f"Kurtosis of {variable2}: {kurtosis2}")
```

```python
#Histogram and Boxplots
import matplotlib.pyplot as plt
import seaborn as sns

histogram_var = 'score_home'

boxplot_var = 'score_away'
filter_var = 'schedule_season'

plt.figure(figsize=(8, 6))
plt.hist(df[histogram_var], bins=10)
plt.xlabel(histogram_var)
plt.ylabel("Frequency")
plt.title(f"Histogram of {histogram_var}")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=filter_var, y=boxplot_var, data=df)
plt.xlabel(filter_var)
plt.ylabel(boxplot_var)
plt.title(f"Box Plots of {boxplot_var} by {filter_var}")
plt.xticks(rotation=45, ha='right')
plt.show()
```
